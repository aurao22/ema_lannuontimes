```python
import urllib
import bs4
#help(bs4)
from urllib import request
import nltk
import time
```

```python
# Etape 1 : se connecter à la page wikipedia et obtenir le code source

def get_page(url):
    req = request.Request(url, headers = {'User-Agent' : 'Mozilla/5.0'})
    html = request.urlopen(req).read()
    return bs4.BeautifulSoup(html, "lxml")
```

```python
actu_html = get_page("https://www.actugaming.net/the-cuphead-show-saison-2-cet-ete-482804/")
actu_html
```

```python
urls = []

for news in news_list:
   urls.append(news.find("a").get("href"))

urls
```

```python
articles = []

for url in urls:
    # Récupération du contenu HTML de l'article
    articles.append(get_page(url).find('div', {'id' : 'wtr-content'}))
```
